```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
acs_cleaned = pd.read_csv('../data/acs_combined.csv')
```

```python
acs_cleaned = acs_cleaned[['year', 'county', 'median_household_income', 'less_than_high_school', 'high_school', 'some_college', 'bachelors_degree', 'graduate_degree', 'total_population']]
```

```python
crimes_cleaned = pd.read_csv('../data/crimes_combined.csv')
```

```python
crimes_cleaned
```

```python
crimes_cleaned = crimes_cleaned[['year', 'county', 'total_crimes', 'crimes_against_persons', 'pct_crimes_against_persons', 'crimes_against_property', 'pct_crimes_against_property', 'crimes_against_society', 'pct_crimes_against_society']]
```

```python
crimes_cleaned
```

join tables

```python
data_merged_df = pd.merge(crimes_cleaned, acs_cleaned, how = 'inner')
```

```python
data_merged_df.info()
```

```python
data_merged_df[['pct_crimes_against_persons', 'pct_crimes_against_property', 'pct_crimes_against_society']] = data_merged_df[['pct_crimes_against_persons', 'pct_crimes_against_property', 'pct_crimes_against_society']].astype('int')
```

```python
data_merged_df['crime_rate_per_1000'] = data_merged_df['total_crimes']/ data_merged_df['total_population'] * 1000
```
